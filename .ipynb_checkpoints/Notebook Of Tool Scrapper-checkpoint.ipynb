```python
import pandas as pd
from selenium import webdriver
import random,time
from urllib.parse import urlparse, parse_qs
```

```python
def openBrowser():
    global driver
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('start-maximized')
    chrome_options.add_argument('start=0')
    chrome_options.add_argument('num=50')
    chrome_options.add_experimental_option('excludeSwitches', ['enable-automation'])
    chrome_options.add_argument('disable-automation')
    chrome_options.add_argument('--incoginto')
    driver = webdriver.Chrome(options=chrome_options)
```

```python
allData = soup.find_all("div",{"class":"g Ww4FFb vt6azd tF2Cxc asEBEc"})
links=[]
for i in allData:
    
    print(allData.index(i)," ",i.text)
    emails = re.findall(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b", str(i))
    print("Emails: ", emails)
#     print(allData.index(i)," ",i.find('div',{'class':'egMi0 kCrYT'}).find('a')['href'])
    url = i.find('div', {'class': 'yuRUbf'}).find('a')['href']
    phone_numbers = re.findall(r"\d{3}-\d{3}-\d{4}", str(i))
    print("Phone numbers: ", phone_numbers)

    follow_links = []
    for a in i.find_all("a", href=True):
        if "facebook" in a.text.lower():
            follow_links.append(a["href"])
    print("Follow us links: ", follow_links)
    print(url)
    links.append(url)
    print("\n")
```

```python
for link in links:
    driver.get(link)
    pageContact=BeautifulSoup(driver.page_source, 'html.parser')
    print(links.index(link)," ",link)
    emails = re.findall(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b", str(pageContact))
    print("Emails: ", emails)
    phone_numbers = re.findall(r"\d{3}-\d{3}-\d{4}", str(pageContact))
    unique_numbers = list(set(phone_numbers))
    print(unique_numbers)

    follow_links = []
    for a in soup.find_all("a", href=True):
        if "follow" in a.text.lower():
            follow_links.append(a["href"])
    print("Follow us links: ", follow_links)
```

```python
response = requests.get("https://www.google.com/search?q=kia dealership colorado&num=100")
soup = BeautifulSoup(response.text, 'html.parser')
allData = soup.find_all("div",{"class":"Gx5Zad fP1Qef xpd EtOod pkphOe"})
```

```python
links=[]
for i in allData:
    
    print(allData.index(i)," ",i.text)
#     print(allData.index(i)," ",i.find('div',{'class':'egMi0 kCrYT'}).find('a')['href'])
    url = i.find('div', {'class': 'egMi0 kCrYT'}).find('a')['href']
    parsed_url = urlparse(url)
    final_url = parse_qs(parsed_url.query)['q'][0]
    print(final_url)
    links.append(final_url)
    print("\n")
```
